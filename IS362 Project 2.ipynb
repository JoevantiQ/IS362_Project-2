Dataset #1: List of best-selling video games from Wikepedia data cleanup & Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv("videogame1_tidy.csv")
```

The first analysis determine which video games have sold the most copies globally, the dataset is sorted by the Global_Sales_Millions column in descending order. The head(10) function is then used to display the top 10 highest-selling games in the dataset.

```python
df.sort_values("Global_Sales_Millions", ascending=False).head(10)
```

The next analysis shows which publishers have generated the most game sales, the dataset is grouped by the Publisher. The total global sales for each publisher are then calculated by summing the Global_Sales_Millions column.

```python
df.groupby("Publisher(s)")["Global_Sales_Millions"].sum().sort_values(ascending=False)
```

The last analysis highlights how many best-selling games appear on each gaming platform. The value_counts() function counts how frequently each platform appears in the Platform(s) column, helping identify the most common platforms among top-selling games.


```python
df["Platform(s)"].value_counts()
```

Overall, each analysis highlights reveail insights into the best-selling video games, the publishers responsible for the highest sales, and the platforms that host the most successful titles. Sorting the dataset helps identify the most popular games, grouping by publishers highlights the companies with the greatest market success, and counting platforms shows which gaming systems are most associated with top-selling titles.

Dataset #2: NBA Scoring Statistics Data Clean-Up & Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv("clean_nbadata.csv")
df.head()
```

The first analysis identifies the players who scored the most total points in the dataset. 
By sorting the dataset by the **PTS (points)** column in descending order, we can determine 
which players contributed the most offensively.

```python
top_scorers = df.sort_values(by="PTS", ascending=False).head(10)

top_scorers[["Player", "Team", "PTS"]]
```

The second analysis focuses on passing ability by identifying the players with the highest number of assists. Assists are an important statistic because they measure a player’s ability to create scoring opportunities for their team.

```python
top_assists = df.sort_values(by="AST", ascending=False).head(10)

top_assists[["Player", "Team", "AST"]]
```

The third analysis examines scoring performance at the team level. Using the groupby() function in pandas, the dataset is grouped by team and the average number of points scored by players on each team is calculated.

```python
team_scoring = df.groupby("Team")["PTS"].mean().sort_values(ascending=False)

team_scoring
```

After cleaning and transforming the dataset, several insights were discovered through  analysis of player statistics. The dataset revealed the top scoring players, the leading  playmakers based on assists, and the teams with the highest average scoring output.

Dataset #3: Box Office Mojo Clean Up & Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv("boxoffice_yearly_clean.csv")

df.head()
```

The first analysis identifies the years with the highest total box office revenue. Sorting the dataset by total gross revenue reveals which years were the most successful financially for the film industry.

```python
highest_gross = df.sort_values(by="Total_Gross", ascending=False).head(5)

highest_gross[["Year","Total_Gross","Top_Movie"]]
```

The second analysis shows which years had the highest number of movie releases. Examining the number of releases helps determine trends in film production over time.

```python
most_releases = df.sort_values(by="Releases", ascending=False).head(5)

most_releases[["Year","Releases"]]
```

The third analysis evaluates which years had the highest average revenue per movie. This metric provides insight into how profitable movies were on average in a given year.

```python
avg_gross = df.sort_values(by="Average_Gross", ascending=False)

avg_gross[["Year","Average_Gross"]].head(10)
```

After cleaning and analyzing the dataset, several trends in the film industry became highlighted. The analysis identified the years with the highest overall box office revenue, the years with the greatest number of film releases, and the years with the highest average revenue per movie. Using pandas made it possible to efficiently organize and analyze the data, demonstrating the importance of properly structured datasets for meaningful analysis.
